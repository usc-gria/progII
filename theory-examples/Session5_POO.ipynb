```python
class Libro:
    def __init__(self,titulo, autor, anho):
        self.titulo = titulo
        self.autor = autor
        self.anho = anho
        self.disponible = None

    def resumen (self):
        return f"{self.titulo} {self.autor} {self.anho}"
    def get_disponible(self):
        return self.disponible
    def set_disponible(self, a):
        self.disponible = a
```

```python
l = Libro("Dune", "Frank Herbert", 2025)
print(l.resumen())
if l.get_disponible():
    print(l.resumen())
else:
    print("El libro no esta disponible")
```

Define la clase Revista con los siguientes atributos:
titulo, numero, anho, disponible, idioma, precio, tematica


```python
class Revista:
    def __init__(self,titulo:str, numero:int, anho, idioma, precio, tematica="Musica", disponible=None):
        self.titulo = titulo
        self.numero = numero
        self.anho = anho
        self.idioma = idioma
        self.precio = precio
        self.tematica = tematica
        self.disponible = disponible

    def actualizar_titulo(self, titulo_nuevo):
        if len(titulo_nuevo) > 100:
            print("El titulo ni puede tener más de 100 ")
        else:
            self.titulo = titulo_nuevo
    def prestar(self):
        if self.disponible:

            self.disponible = False
            return True
        else:
            return False
    def resumen(self):
        return f"{self.titulo} #{self.numero} ({self.anho}) [{self.idioma}] {self.precio}€ {self.tematica}"
    def __str__(self):
        return f"{self.titulo} #{self.numero} ({self.anho}) [{self.disponible}]"
    def __repr__(self):
        return f"Revista(titulo={self.titulo!r}, numero={self.numero!r}, anho={self.anho!r})"
```
