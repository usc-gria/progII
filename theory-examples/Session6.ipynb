# Ejemplos POO y Encapsulación

```python
class Revista:

    def __init__(self, titulo, autor):
        self.__titulo = titulo
        self.__autor = autor
        self.__precio = 0

    def get_titulo(self):
        return self.__titulo
    def set_titulo(self, nuevo_titulo:str):
        if type(nuevo_titulo) == str and len(nuevo_titulo)>0:
            self.__titulo = nuevo_titulo
        else:
            raise Exception("El valor no es una cadena de caracteres válida")


    def set_precio(self, nuevo_precio):
        if (type(nuevo_precio) == float) and ( nuevo_precio>=0):
            self.__precio = nuevo_precio
```

```python
r1 = Revista("Hola", "David")
r1.titulo = True
print(r1.titulo)
```

```python
r2 = Revista("Hola", "David")
print(r2.get_titulo())
try:
    r2.set_titulo("Adios")
except Exception as e:
    print(e)
print(r2.get_titulo())
```

### Ejemplo "moderno" de setter y getter

```python
class Revista:

    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor

    @property
    def titulo(self):
        return self._titulo

    @titulo.setter
    def titulo(self, nuevo_titulo:str):
        if isinstance(nuevo_titulo,str) and len(nuevo_titulo)>0:
            self._titulo = nuevo_titulo
        else:
            raise Exception("El valor no es una cadena de caracteres válida")
```

```python
r1 = Revista("hola", "david")
print(r1.titulo)
```
