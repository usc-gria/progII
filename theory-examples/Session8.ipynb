# Ejemplos POO y Herencia

```python
class Item:

    def __init__(self, titulo, autor, anho):
        self.titulo = titulo
        self.autor = autor
        self.anho = anho

    @property
    def titulo(self):
        return self._titulo

    @titulo.setter
    def titulo(self, nuevo_titulo:str):
        if isinstance(nuevo_titulo,str) and len(nuevo_titulo)>0:
            self._titulo = nuevo_titulo
        else:
            raise Exception("El valor no es una cadena de caracteres válida")

    @property
    def autor(self):
        return self._autor

    @autor.setter
    def autor(self, nuevo_autor:str):
        if isinstance(nuevo_autor,str) and len(nuevo_autor)>0:
            self._autor = nuevo_autor
        else:
            raise Exception("El valor no es una cadena de caracteres válida")
    def resumen(self):
        return f"{self.titulo} {self.autor} {self.anho}"
```

```python
class Revista(Item):

    def __init__(self, titulo, autor, anho, numero):
        super().__init__(titulo,autor,anho)
        self.numero = numero

    def resumen(self):
        return f"{super().resumen()} - {self.numero}"
```

```python
r1 = Revista("Hola","Autor",2025,1)
print(r1.resumen())
```
