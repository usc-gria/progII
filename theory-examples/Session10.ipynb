```python
def factorial(n:int) -> int:
    if n == "a":
        return 1
    print(f"Estoy en el factorial de {n}")
    return n * factorial(n-1)

print(factorial(5))
```

```python
def busqueda_binaria(lista, x, inicio, fin): # lista debe estar ordenada
    if inicio > fin:
        return -1
    medio = (inicio + fin) // 2 # // devuelve la división sin decimales
    if lista[medio] == x:
        return medio
    elif x < lista[medio]:
        return busqueda_binaria(lista, x, inicio, medio - 1)
    else:
        return busqueda_binaria(lista, x, medio + 1, fin)


mi_lista = list(range(1,5000))
busqueda_binaria(mi_lista, 678, 0, len(mi_lista)-1)
```

# Árboles Binarios

```python
class NodoAB:
    def __init__(self, val, izq=None, der=None):
        self.valor = val
        self.hijoIzq = izq
        self.hijoDer = der

    @property
    def valor(self):
        return self._valor

    @valor.setter
    def valor(self, val):
        self._valor = val

    @property
    def hijoIzq(self):
        return self._hijoIzq

    @hijoIzq.setter
    def hijoIzq(self, nodo):
        self._hijoIzq = nodo

    @property
    def hijoDer(self):
        return self._hijoDer

    @hijoDer.setter
    def hijoDer(self, nodo):
        self._hijoDer = nodo

    def __str__(self):
        return str(self.valor)

    def insertIzq(self, nuevoval):
        if self.hijoIzq is None:
            self.hijoIzq = NodoAB(nuevoval)
        else:
            nuevoNodo = NodoAB(nuevoval)
            nuevoNodo.hijoIzq = self.hijoIzq
            self.hijoIzq = nuevoNodo

    def insertDer(self, nuevoval):
        if self.hijoDer is None:
            self.hijoDer = NodoAB(nuevoval)
        else:
            nuevoNodo = NodoAB(nuevoval)
            nuevoNodo.hijoDer = self.hijoDer
            self.hijoDer = nuevoNodo

    def tamanho(self):
        tam = 1
        if self.hijoIzq is not None:
            tam += self.hijoIzq.tamanho()
        if self.hijoDer is not None:
            tam += self.hijoDer.tamanho()
        return tam

    def dibuja(self, prefijo):
        if self.hijoDer is not None:
            self.hijoDer.dibuja(prefijo + " ")
        print(prefijo + "|-- " + str(self.valor))
        if self.hijoIzq is not None:
            self.hijoIzq.dibuja(prefijo + " ")

    def preorden(self):
        print(self, end=" ")
        if self.hijoIzq is not None:
            self.hijoIzq.preorden()
        if self.hijoDer is not None:
            self.hijoDer.preorden()

    def listaPreorden(self, listaPrevia):
        listaPrevia.append(self.valor)
        if self.hijoIzq is not None:
            self.hijoIzq.listaPreorden(listaPrevia)
        if self.hijoDer is not None:
            self.hijoDer.listaPreorden(listaPrevia)
        return listaPrevia
```
