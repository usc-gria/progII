# Ejemplos POO y Composicion

```python
class Libro:

    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor

    @property
    def titulo(self):
        return self._titulo

    @titulo.setter
    def titulo(self, nuevo_titulo:str):
        if isinstance(nuevo_titulo,str) and len(nuevo_titulo)>0:
            self._titulo = nuevo_titulo
        else:
            raise Exception("El valor no es una cadena de caracteres válida")

    @property
    def autor(self):
        return self._autor

    @autor.setter
    def autor(self, nuevo_autor:str):
        if isinstance(nuevo_autor,str) and len(nuevo_autor)>0:
            self._autor = nuevo_autor
        else:
            raise Exception("El valor no es una cadena de caracteres válida")
```

```python
class Biblioteca:
    def __init__(self, libros:list):
        self.libros = libros

    @property
    def libros(self):
        return self._libros

    @libros.setter
    def libros(self, nuevos_libros:list):
        if len(nuevos_libros) >0:
            flag = True
            for mi_libro_nuevo in nuevos_libros:
                if not isinstance(mi_libro_nuevo,Libro):
                    flag = False
            if flag:
                self._libros = nuevos_libros
            else:
                raise Exception("Los objetos deben ser libros")

    def add_book(self, nuevo_libro:Libro):
        if isinstance(nuevo_libro,Libro):
            self.libros.append(nuevo_libro)
```

```python
l1 = Libro("Titulo", "autor")
l2 = Libro("L2", "autor2")

mis_libros = [l1, l2]
b = Biblioteca(mis_libros)

for libro in b.libros:
    print(libro.titulo)
```
