```python
import time

def suma_n(n):  # algoritmo A
  inicio = time.time()  # tiempo de inicio del ordenador (s)
  suma = 0
  for i in range(1, n + 1):
    suma += i
  fin = time.time()  # tiempo de parada
  return suma, fin - inicio

print("tiempo para n = ", 100000)
for t in range(1,6):
    print("Suma es %d en %.7f segundos" % (suma_n(100000)))
```

```python
import time

def suma_n(n):  # algoritmo A
  inicio = time.time()  # tiempo de inicio del ordenador (s)
  suma = 0
  for i in range(1, n + 1):
    suma += i
  fin = time.time()  # tiempo de parada
  return suma, fin - inicio

print("\ntiempos para n = 10,000, 100,000,1,000,000, 100,000,000")
for t in [10000, 100000, 1000000, 100000000]:
    print( "Suma %10d números en %10.7f segundos"%(suma_n(t)))
```

```python
def suma2_n(n):
  inicio = time.time()
  suma = n*(n+1)/2
  fin = time.time()
  return suma, fin-inicio

print("\ntiempos para n = 10,000, 100,000,1,000,000, 100,000,000")
for t in [10000, 100000, 1000000, 100000000]:
    print( "Suma %10d números en %10.7f segundos"%(suma2_n(t)))
```
