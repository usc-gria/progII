```python
import time

def suma_n(n):
    suma = 0
    inicio = time.time()
    for i in range(1,n+1):
        suma += i
    final = time.time()
    return suma, final-inicio
print("--- Ejecución Algoritmo 1------")
valores = [10, 100, 1000, 10000, 1000000, 10000000, 100000000]
for n in valores:
    resultado, tiempo = suma_n(n)
    print(f"Suma es {resultado} en {tiempo} segundos")


def suma2_n(n):
    suma = 0
    inicio = time.time()
    for i in range(1,n+1):
        termino = i
        suma = suma + termino
    final = time.time()
    return suma, final-inicio
print("--- Ejecución Algoritmo 2------")
valores = [10, 100, 1000, 10000, 1000000, 10000000, 100000000]
for n in valores:
    resultado, tiempo = suma2_n(n)
    print(f"Suma es {resultado} en {tiempo} segundos")
```

```python
    #!/usr/bin/env python3
"""
demo_explosive_complexities.py

Demostración didáctica de por qué O(2^n) (subconjuntos) y O(n!) (permutaciones)
“explotan”, midiendo tiempos con perf_counter.

- O(2^n): generamos TODOS los subconjuntos (sin guardarlos todos, para no petar memoria).
- O(n!): iteramos TODAS las permutaciones (tampoco las guardamos), sólo las contamos.

Ejecuta:
    python demo_explosive_complexities.py
"""

from __future__ import annotations

from itertools import combinations, permutations
from time import perf_counter
from typing import Iterable, Callable, Any


# -------------------------
# Algoritmos / generadores
# -------------------------

def iter_all_subsets(items: list[int]) -> Iterable[tuple[int, ...]]:
    """
    Genera todos los subconjuntos (combinaciones de cualquier tamaño).
    Total: 2^n subconjuntos.
    """
    n = len(items)
    for r in range(n + 1):
        yield from combinations(items, r)


def count_subsets(items: list[int]) -> int:
    """Cuenta subconjuntos iterándolos (O(2^n) tiempo, O(1) memoria adicional)."""
    c = 0
    for _ in iter_all_subsets(items):
        c += 1
    return c


def count_permutations(items: list[int]) -> int:
    """Cuenta permutaciones iterándolas (O(n!) tiempo, O(1) memoria adicional)."""
    c = 0
    for _ in permutations(items):
        c += 1
    return c


# -------------------------
# Timing utils
# -------------------------

def time_call(fn: Callable[[], Any], repeats: int = 1) -> float:
    """
    Devuelve tiempo medio (en segundos) de ejecutar fn() 'repeats' veces.
    """
    t0 = perf_counter()
    for _ in range(repeats):
        fn()
    t1 = perf_counter()
    return (t1 - t0) / repeats


def fmt_s(seconds: float) -> str:
    if seconds < 1e-6:
        return f"{seconds * 1e9:.1f} ns"
    if seconds < 1e-3:
        return f"{seconds * 1e6:.1f} µs"
    if seconds < 1:
        return f"{seconds * 1e3:.1f} ms"
    return f"{seconds:.3f} s"


# -------------------------
# Experimentos
# -------------------------

def run_experiment_2_pow_n(ns: list[int]) -> None:
    print("\n=== O(2^n): contar subconjuntos ===")
    print("n | 2^n (subconjuntos) | tiempo")
    print("--+---------------------+----------------")
    for n in ns:
        items = list(range(n))

        expected = 2 ** n

        # pocas repeticiones porque esto sube rápido
        repeats = 3 if n <= 18 else 1

        elapsed = time_call(lambda: count_subsets(items), repeats=repeats)
        print(f"{n:2d} | {expected:19d} | {fmt_s(elapsed)}")


def factorial(n: int) -> int:
    f = 1
    for k in range(2, n + 1):
        f *= k
    return f


def run_experiment_factorial(ns: list[int]) -> None:
    print("\n=== O(n!): contar permutaciones ===")
    print("n | n! (permutaciones) | tiempo")
    print("--+---------------------+----------------")
    for n in ns:
        items = list(range(n))

        expected = factorial(n)

        # factorial crece brutal: cuidado con n > 11/12 (depende de tu máquina)
        repeats = 3 if n <= 9 else 1

        elapsed = time_call(lambda: count_permutations(items), repeats=repeats)
        print(f"{n:2d} | {expected:19d} | {fmt_s(elapsed)}")


def main() -> None:
    print("Demostración de complejidades explosivas (2^n y n!)\n"
          "Nota: los tiempos dependen mucho de tu CPU.\n"
          "Consejo: sube n poco a poco si lo ejecutas en clase.\n")

    # Valores razonables para que esto termine en una máquina normal.
    # Ajusta en directo según veas los tiempos.
    ns_2pow = [5, 10, 15, 18, 20, 22]   # 2^22 ~ 4.2M (puede tardar ya)
    ns_fact = [5, 7, 8, 9, 10, 11]      # 11! ~ 39.9M (puede tardar bastante)

    run_experiment_2_pow_n(ns_2pow)
    run_experiment_factorial(ns_fact)

    print("\nInterpretación rápida:")
    print("- En 2^n, cada elemento duplica el número de posibilidades (coger/no coger).")
    print("- En n!, el número de órdenes posibles crece multiplicando n*(n-1)*...*1.")
    print("- Si en clase quieres 'impacto visual', prueba a subir n en +1 y ver el salto.")


if __name__ == "__main__":
    main()
```
