```python
import time

def suma_n(n):
    suma = 0
    inicio = time.time()
    for i in range(1,n+1):
        suma += i
    final = time.time()
    return suma, final-inicio
print("--- Ejecución Algoritmo 1------")
valores = [10, 100, 1000, 10000, 1000000, 10000000, 100000000]
for n in valores:
    resultado, tiempo = suma_n(n)
    print(f"Suma es {resultado} en {tiempo} segundos")


def suma2_n(n):
    suma = 0
    inicio = time.time()
    for i in range(1,n+1):
        termino = i
        suma = suma + termino
    final = time.time()
    return suma, final-inicio
print("--- Ejecución Algoritmo 2------")
valores = [10, 100, 1000, 10000, 1000000, 10000000, 100000000]
for n in valores:
    resultado, tiempo = suma2_n(n)
    print(f"Suma es {resultado} en {tiempo} segundos")
```

## Algoritmo B

```python
import time

def suma2_n(n):
  inicio = time.time()
  suma = n*(n+1)/2
  fin = time.time()
  return suma, fin-inicio

print("\ntiempos para n = 10,000, 100,000,1,000,000, 100,000,000")
for t in [10000, 100000, 1000000, 100000000]:
    print( "Suma %10d números en %10.15f segundos"%(suma2_n(t)))
```
