# Listas Enlazadas

```python
class Nodo:

    def __init__(self, dato, sig=None):
        self.dato = dato
        self.sig = sig

    @property
    def dato(self):
        return self._dato

    @dato.setter
    def dato(self, dato):
        self._dato = dato

    @property
    def sig(self) -> Nodo:
        return self._sig

    @sig.setter
    def sig(self,nodo):
        self._sig = nodo

    def __str__(self):
        return str(self._dato)
    def __repr__(self):
        return f"Nodo(dato={self._dato!r})"
```
