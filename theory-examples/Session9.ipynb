# Listas Enlazadas

```python
class Nodo:

    def __init__(self, dato, sig=None):
        self.dato = dato
        self.sig = sig

    @property
    def dato(self):
        return self._dato

    @dato.setter
    def dato(self, dato):
        self._dato = dato

    @property
    def sig(self) -> Nodo:
        return self._sig

    @sig.setter
    def sig(self,nodo):
        self._sig = nodo

    def __str__(self):
        return str(self._dato)
    def __repr__(self):
        return f"Nodo(dato={self._dato!r})"
```

```python

class ListaEnlazada:
    def __init__(self, dato=None):
        self.cabeza = Nodo(dato) if dato is not None else None
        self._tamano = 1 if dato is not None else 0

    @property
    def cabeza(self):
        return self._cabeza

    @cabeza.setter
    def cabeza(self, nodo):
        if nodo is not None and not isinstance(nodo, Nodo):
            raise TypeError("cabeza debe ser un Nodo o None")
        self._cabeza = nodo

    @property
    def tamano(self):
        return self._tamano

    def anade_elemento_cabeza(self, dato):
        if self.cabeza is None:
            self.cabeza = Nodo(dato)
        else:
            siguiente = self.cabeza
            self.cabeza = Nodo(dato, siguiente)
        self._tamano += 1

    def extrae_elemento_cabeza(self):
        if self.cabeza is None:
            raise IndexError("La lista está vacía")
        dato = self.cabeza.dato
        self.cabeza = self.cabeza.sig
        self._tamano -= 1
        return dato

    def to_list(self) -> list:
        valores = [self.cabeza.dato] if self.cabeza is not None else []
        nodo_siguiente = self.cabeza.sig
        while nodo_siguiente is not None:
            valores.append(nodo_siguiente.dato)
            nodo_siguiente = nodo_siguiente.sig
        return valores

    def __iter__(self):
        actual = self.cabeza
        while actual is not None:
            yield actual
            actual = actual.sig
```

```python
class Pila(ListaEnlazada):
    def __init__(self, dato=None):
        super().__init__(dato)

    def push(self, dato):
        self.anade_elemento_cabeza(dato)

    def pop(self):
        return self.extrae_elemento_cabeza()

    def peek(self):
        if self.cabeza is None:
            raise IndexError("pila vacía")
        return self.cabeza.dato

    def is_empty(self):
        return self.tamano == 0
```

```python
pila = Pila()
pila.push(1)
pila.push(2)
pila.push(3)

for valor in pila:
    print(valor)
```

```python
def parentesis_balanceados(s: str) -> bool:
    pila = Pila()

    for ch in s:
        if ch == "(":
            pila.push(ch)
        elif ch == ")":
            if pila.is_empty():
                return False
            pila.pop()

    return pila.is_empty()
```

```python
assert parentesis_balanceados("((a+b)*c)") is True
assert parentesis_balanceados("(a+b))") is False
assert parentesis_balanceados("(()") is False
assert parentesis_balanceados(")(") is False
assert parentesis_balanceados(")(") is True
```

```python
class Cola(ListaEnlazada):

    def __init__(self, dato=None):
        super().__init__(dato)

    def add(self, dato):
        nuevo = Nodo(dato)

        if self.cabeza is None:
            self.cabeza = nuevo
        else:
            actual = self.cabeza
            while actual.sig is not None:
                actual = actual.sig
            actual.sig = nuevo

        self._tamano += 1

    def first(self):
        if self.cabeza is None:
            raise IndexError("La cola está vacía")
        return self.cabeza.dato

    def remove(self):
        if self.cabeza is None:
            raise IndexError("La cola está vacía")

        dato = self.cabeza.dato
        self.cabeza = self.cabeza.sig
        self._tamano -= 1
        return dato

```
